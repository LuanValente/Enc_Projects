```python
import pandas as pd #saque sempre com todos os seguros (Sem seguro,Super,Master)(Todos Os Seguros)
import numpy as np #Venda (Todos os produtos)}
import re #Ativação (Todos os produtos)}
```

```python
df = pd.read_excel("bmg2 (2).xlsx")
```

```python
df.head()
```

```python
df = df.iloc[2:]

# Redefina os índices após a remoção da primeira linha
df.reset_index(drop=True, inplace=True)
df.head()
```

```python

# Definindo a primeira linha como cabeçalhos
df.columns = df.iloc[0]

# Removendo a primeira linha dos dados
df = df[1:]
df.head()
```

```python
# Drop da coluna 'Cidade'
df.reset_index(inplace=True)
df = df.drop(columns=['Nova comissão diferida'])
# Criando a coluna 'Tabela' com a concatenação de 'Logo' e 'Descrição Entidade'
df['Tabela'] = df['Logo'].astype(str) + ' - ' + df['Descrição Entidade']
df.head()
```

```python
# Renomeando a coluna 'Antecipação Equivalente Liquida' para 'saque'
df = df.rename(columns={'Antecipação Equivalente Liquida ': 'Saque'})
df.head()
```

```python
saque_sem_seg = df[['Tabela']].copy()
saque_sem_seg['Tabela'] = saque_sem_seg['Tabela'] + ' (Seguro Prestamista)'
saque_sem_seg['Saque'] = df['Saque']
saque_sem_seg.head()
from IPython.display import display, FileLink
# Salvar a tabela concatenada em um arquivo Excel
nome_arquivo = 'Sem Seguro.xlsx'
saque_sem_seg.to_excel(nome_arquivo, index=False)

# Mostrar o botão de download
display(FileLink(nome_arquivo, result_html_prefix="Clique para baixar: "))
```
